````python
import matplotlib.pyplot as plt

# Vector 정리
x = [2,-1]
y = [1,2]

# 2D grape 생성
plt.figure()

# 벡터 위치 플로팅
plt.quiver(0,0,x[0],y[0],angles='xy',scale_units='xy',scale=1,color='r',label='Vector(2,1)')
plt.quiver(0,0,x[1],y[1],angles='xy',scale_units='xy',scale=1,color='b',label='Vector(-1,2)')

# 축 범위 설정
plt.xlim(-2,3)
plt.ylim(-2,3)

# 축 레이블 및 타이틀 설정
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('2-D Vector plot')

# 범례 표시
plt.legend()

# 그래프 출력
plt.grid(True)
plt.show()
````

### 만약 결과값이 (0,3) 이라면?
- x가 1이고 y가 2일 경우 성립된다.

그래프로 설명한다면 빨간 화살표의 끝점에서 y(-1,2)만큼 두번 이동하게 된다면 (0,3)에서 끝날 것이다.

````python
# 필요한 라이브러리 임포트
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 벡터 정의
x = [2, -1, 0]
y = [-1, 2, -3]
z = [0, -1, 4]

# 3D 그래프 생성
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 벡터 위치 플로팅
ax.quiver(0, 0, 0, x[0], y[0], z[0], color='r', label='Vector 1 (2,-1,0)')
ax.quiver(0, 0, 0, x[1], y[1], z[1], color='g', label='Vector 2 (-1,2,-3)')
ax.quiver(0, 0, 0, x[2], y[2], z[2], color='b', label='Vector 3 (0,-1,4)')

# 축 설정
ax.set_xlim([-2, 3])
ax.set_ylim([-3, 3])
ax.set_zlim([-4, 5])

# 레이블 및 타이틀 설정
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3-D Vector Plot')

# 범례 표시
ax.legend()

# 그래프 출력
plt.show()
````

# 행렬 곱셈

## Ax = b
- Ax is a combination of columns of A

````python
import numpy as np

# 2x2 행렬 A
A = np.array([[2, 5],
              [1, 3]])

# 2x1 벡터 b
b = np.array([[1],
              [2]])

# 행렬 곱 Ax = b
result = np.dot(A, b)  # 또는 A @ b

# 결과 출력
print("A * b = \n", result)

````
